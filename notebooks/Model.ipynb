```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
import csv
```

```python
# Split train.csv to x_train and y_train
train_df = pd.read_csv('data/train.csv')
train_numpy = train_df.to_numpy()

x_train = train_numpy[:,:-1]
y_train = train_numpy[:,12]

# Read test.csv to x_test
test_df = pd.read_csv('data/test.csv')
x_test = test_df.to_numpy()

print('x_train shape:', x_train.shape) # (2183910, 12)
print('y_train shape:', y_train.shape) # (2183910,)
print('x_test shape:', x_test.shape) # (106692, 12)
```

```python
# List of Regressors
rf = RandomForestRegressor(n_estimators=10, random_state=1)
kn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state = 0)
xgb = XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, alpha = 10, n_estimators = 10)

vot = VotingRegressor([('rf', rf), ('kn', kn), ('dt', dt), ('xgb', xgb)])
vot.fit(x_train, y_train) # Train Model
y_pred = vot.predict(x_test) # Predict

print("Number of Predictions: ", len(y_pred))

# Write predictions to file
predictions = zip(range(len(y_pred)), y_pred)
with open("data/submission.csv","w") as pred:
    csv_out = csv.writer(pred)
    csv_out.writerow(['id','predicted'])
    for row in predictions:
        csv_out.writerow(row)

print("Predictions written to file")
```
